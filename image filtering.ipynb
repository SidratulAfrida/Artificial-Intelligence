```python
import cv2
import numpy as np
from matplotlib import pyplot as plt


image = cv2.imread('image.jpeg', cv2.IMREAD_GRAYSCALE)

def apply_arithmetic_mean_filter(image, filter_size):
    kernel = np.ones((filter_size, filter_size), np.float32) / (filter_size ** 2)
    result = cv2.filter2D(image, -1, kernel)
    return result


filtered_image_3x3 = apply_arithmetic_mean_filter(image, 3)
filtered_image_7x7 = apply_arithmetic_mean_filter(image, 7)
filtered_image_9x9 = apply_arithmetic_mean_filter(image, 9)

plt.subplot(2, 2, 1), plt.imshow(image, cmap='gray'), plt.title('Original')
plt.subplot(2, 2, 2), plt.imshow(filtered_image_3x3, cmap='gray'), plt.title('3x3 Filter')
plt.subplot(2, 2, 3), plt.imshow(filtered_image_7x7, cmap='gray'), plt.title('7x7 Filter')
plt.subplot(2, 2, 4), plt.imshow(filtered_image_9x9, cmap='gray'), plt.title('9x9 Filter')

plt.show()
```
